```
import os, cv2
from PIL import Image, ImageOps
from tqdm import tqdm
import numpy as np
import math
import timeit
from random import shuffle
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.models import Sequential, Model
from keras.layers import Dense, Activation, Dropout, Flatten, BatchNormalization
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras import backend as K
from keras.optimizers import Adam
from keras.utils import to_categorical
from sklearn.metrics import confusion_matrix

import warnings
warnings.filterwarnings('ignore')

os.environ['KMP_DUPLICATE_LIB_OK']='True' # to fix kernel dead and restarting issue

%matplotlib inline
```

```
dataset_path = '/content/drive/My Drive/Academics/Sem1/682/CS682 Project/Milestone/Crop_Data/'
```

```
data_list = ['Background', 'Fish', 'Lamprey']
preprocessing = False

X = []
X_test = []
y = []
for i, data in enumerate(data_list):
  # print(i, data)
  files = os.listdir(dataset_path + data)
  # print('len', len(files))
  for j, file in enumerate(files):
    if j%200 == 0:
      print(i, j)
    if file == '.DS_Store':
      pass
    else:
      # print('Image path:', data_dir + data + '/' + file)
      # print(i, j)
      img = cv2.imread(dataset_path + data + '/' + file)
      img = img / 255.0
      X.append(img)
      y.append(i)

X = np.array(X)
y = np.array(y)
idxs = np.arange(len(X))
# shuffle(idxs)
y = y[shuffle(np.arange(len(X)))]
num_classes = 3
y = np.eye(num_classes)[y]
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 682)

print("Size of training set : ", len(X_train))
print("Size of test set : ", len(X_test))
print("X_train shape : ", X_train.shape, end = '; ')
print("y_train shape : ", y_train.shape)
print("X_test shape : ", X_test.shape, end = '; ')
print("y_test shape : ", y_test.shape)
```
